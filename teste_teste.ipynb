```python
from pathlib import Path
import pandas as pd 
input_dir = Path.cwd() / 'DOWNLOADS_FECHAMENTO'

fechamento = list(input_dir.glob("*.xls*"))


print (fechamento)
```

```python
#Armazenamento dos dataframes em uma lista
lista_fechamento = []

for path in list(input_dir.glob('*.xls*')):
    part = pd.read_excel(path)
    lista_fechamento.append(part)
```

```python
#concatenação dos df

df = pd.concat(lista_fechamento)

#salvar o df no diretório
output_dir = Path.cwd() / 'Fechamento_Geral'
output_dir.mkdir(exist_ok=True)
df.to_excel(output_dir / 'Fechamento_Geral.xlsx' , index=False) 

```
